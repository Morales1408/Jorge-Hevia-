### Calculo el p valor entre las respuestas puestas en el  ndc y en el beck








No había calculado antes el p valor de esta manera, estuve leyendo y según yo, es correcto en cómo lo hice, sin embargo, dejo aquí el cómo y por qué funciona
https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html

```python
#Creo una función, para que lo estipulado arriba sea muchísimo más fácil y rápido 

def beckvsndc1(beck,lista):
    tbeck = 'beck:'+str(beck)
    for i in range(1,19):
        ndc = i
        tndc = 'ndc:'+str(ndc)
        ndc_v = data[tndc].tolist()
        beck_v = data[tbeck].tolist()
        slope, intercept, r_value, p_value, std_err = stats.linregress(ndc_v,beck_v)
        print('Significacncia estadística entre la pregunta #'+str(beck),'del beck y la pregunta #'+str(ndc),'del ndc',p_value,'\n')
        if p_value <= 0.05:
            texto = 'La pregunta del beck #'+str(beck),'con la pregunta #'+str(i),'del ndc'
            lista.append(texto)
        #print(lista)
    return lista
```
