```python
#Librerías que me sirvieron
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML, display_html, display
```

```python
#Aquí subo la primera olas de resultados
data1 = pd.read_csv('data.csv',index_col = False)
#Muestro la información
data1.head()
```

```python
#Aquí subo la segunda ola de respuestas
data2 = pd.read_csv('nuevo.csv',index_col = False)
#Muestro la informacipn
data2.head()
```

Debido a que la primera ola de respuestas carece de resultados en la columna "Cuarentena", y además, no tiene ninguna entrada en cuanto a resultados, las eliminaré del segundo dataset para poder continuar

```python
#Elimino las tres columnas
data2 = data2.drop(['cuarentena_1:1'],axis=1)
data2 = data2.drop(['cuarentena_2:1'],axis=1)
data2 = data2.drop(['cuarentena_3:1'],axis=1)
#Muestro la información
data2.head()
```

Ahora que tengo ambos datasets con el mísmo número de columnas, uno ambos datasets para tener uno solo sobre el cuál trabajar

```python
#uno los archivos
data = data1.append(data2)
#Reinicio el index, porque al mezclarse, se confunden
data = data.reset_index()
#Elimino la columna "Index" que se creará automático
data = data.drop(['index'],axis = 1)
#Muestro la información resultante
data.tail()
```

Ahora que tengo tan solo un dataset con toda la información, eliminaré a todos aquellos participantes que dejaron una o más preguntas sin contestar

```python
#Elimino entradas vacías 
data = data.dropna()
#Reinicio el index porque al eliminar las entradas vacías ya no coincidirán
data = data.reset_index()
#Elimino la columna index que se creará automáticamente, pero no sirve de nada
data = data.drop(['index'],axis = 1)
#Muestro la información 
data.head()
```

Como medidas de optimización, eliminaré todas las columnas que no me son de utilidad para responder la pregunta

```python
#Elimino columnas que no me servirán
data = data.drop(['participant'],axis = 1)
data = data.drop(['age:1'],axis = 1)
data = data.drop(['noticias:1'],axis = 1)
data = data.drop(['igt:1'],axis = 1)
data = data.drop(['previo:1'],axis = 1)
data = data.drop(['previo:2'],axis = 1)
data = data.drop(['previo:3'],axis = 1)
data = data.drop(['previo:4'],axis = 1)
data = data.drop(['proximo:1'],axis = 1)
data = data.drop(['country'],axis = 1)
```

Ahora que he hecho lo básico en cuanto al preprocesamiento de la información, procedo a calcular el promedio de la prueba de beck y del NDC para luego poder determinar si las personas tienen un alto o bajo nivel y poder responder posteriormente las preguntas

```python
#Calculo el promedio de ndc por alumno
data['promedio ndc'] = (data['ndc:1']+data['ndc:2']+data['ndc:3']+data['ndc:4']+data['ndc:5']+data['ndc:6']+data['ndc:7']+
                       data['ndc:8']+data['ndc:9']+data['ndc:10']+data['ndc:11']+data['ndc:12']+
                       data['ndc:13']+data['ndc:14']+data['ndc:15']+data['ndc:16']+
                       data['ndc:17']+data['ndc:18'])/18
#Calculo el promedio de la prueba de beck por alumno
data['promedio beck'] =(data['beck:1']+data['beck:2']+data['beck:3']+data['beck:4']+data['beck:5']+data['beck:6']+data['beck:7']+
                       data['beck:8']+data['beck:9']+data['beck:10']+data['beck:11']+data['beck:12']+
                       data['beck:13']+data['beck:14']+data['beck:15']+data['beck:16']+
                       data['beck:17']+data['beck:18']+data['beck:19']+data['beck:20']+data['beck:21'])/21
```

Ahora que conozco los promedios de ambos puntajes (ndc y beck), proceso a determinar si las personas obtuvieron un puntaje alto o un puntaje bajo en el NDC. Lo que recomendó usted fué calcular la mediana y utilizarla como treshold

```python
#Calculo la mediana de los promedios del ndc 
mediana = data['promedio ndc'].median()
#Creo una columna para poder guardar los resultados
data['estado'] = 0
#Clasifico los puntajes del ndc por si son bajos o altos
for index, col in data.iterrows():
    if data['promedio ndc'][index] > mediana:
        data['estado'][index] = 'Alto'
    elif data['promedio ndc'][index] < mediana:
        data['estado'][index] = 'Bajo'
```

No se asuste por lo de arriba, es una advertencia de python por correr y manipular datasets de manera "ineficiente"

```python
#Para una mejor visualización, eilimno ahora todas las columnas que me serán irrelevantes
data = data[['gender:1','promedio ndc','promedio beck','TIME_start','TIME_end','TIME_total','estado']]
#Muestro mi información
data.head()
```

```python
#Ahora voy a sustituir los valores booleanos por el significado en la columna de genero
data['gender:1'] = data['gender:1'].replace(1,'hombre')
data['gender:1'] = data['gender:1'].replace(2,'mujer')
#Muestro mi informacion
data.head()
```

Ahora divido los datos. Crearé un dataset por cada conbinación posible, es decir, crearé los siguientes datasets:

* Mujeres con puntajes altos (m_a)
* Mujeres con puntajes bajos (m_b)
* Hombres con puntajes bajos (h_b)
* Hombres con puntajes altos (h_a)

```python
#Hago la separación de los datos
m_a = data[(data['gender:1'] == 'mujer') & (data['estado']=='Alto')]
m_b = data[(data['gender:1'] == 'mujer') & (data['estado']=='Bajo')]
h_a = data[(data['gender:1'] == 'hombre') & (data['estado']=='Alto')]
h_b = data[(data['gender:1'] == 'hombre') & (data['estado']=='Bajo')]
```

Aquí creo la tabla que pidió. Por practicidad, cree dos tablas, una para los valores altos y otra para los bajos. Si quiere puedo hacer una sola como en su ejemplo, pero como los datos no son del mismo tamaño (no es el mísmo número de personas con puntajes bajos que altos), la tabla queda fea

```python
altos = m_a[['gender:1','promedio beck']].append(h_a[['gender:1','promedio beck']])
display(HTML('<h2>Personas con un puntaje alto en el NDC</h2>'))
display_html(altos)
bajos = altos = m_b[['gender:1','promedio beck']].append(h_b[['gender:1','promedio beck']])
display(HTML('<h2>Personas con un puntaje bajo en el NDC</h2>'))
display_html(bajos)
```

```python
#Ordeno los valores de manera ascendenete 
m_a = m_a.sort_values('promedio ndc',ascending=True)
h_a = h_a.sort_values('promedio ndc',ascending=True)
```

## Visualizando la información

```python
plt.figure(figsize=[10,10])
plt.xlabel('Promedio NDC',fontsize=15)
plt.ylabel('Promedio Beck',fontsize=15)
plt.title('Relación entre una calificación alta en mujeres en el NDC y la prueba Beck',fontsize=15)
#plt.plot(h_a['promedio ndc'],h_a['promedio beck'],'ko',color='blue',label = 'hombres')
plt.plot(m_a['promedio ndc'],m_a['promedio beck'],'ko',color='red',label = 'mujeres')
plt.savefig('ndcvsbeck.jpg',dpi=500)
plt.legend()
plt.show()
```
